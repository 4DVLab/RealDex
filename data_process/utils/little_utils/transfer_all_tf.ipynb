```python
import os
import numpy as np
import shutil
from pathlib import Path

def transfer_all_bags_tf_folder(tf_search_folder,tf_save_folder,post):
    #you should scan all the folder, then come in the folder, 
    #if don't do this, you would scan all the images first 
    now_folder = tf_search_folder / post
    for file in os.listdir(now_folder):
        temp_post = post / file
        now_path = tf_search_folder / temp_post
        save_path = tf_save_folder / post #不要TF在里面
        print(now_path)
        try:
            if os.path.isdir(now_path):
                if "TF" in file:
                    os.makedirs(save_path,exist_ok=True)
                    # os.system(f"cp -r {now_path} {save_path}")
                    # os.system(f"cp -r {now_folder}/global_name_position {save_path}")
                    os.system(f"cp -r {now_folder}/cam0/rgb {save_path}")
                    # shutil.copy2(now_path,save_path)
                    return
                else:
                    transfer_all_bags_tf_folder(tf_search_folder,tf_save_folder,temp_post)
        except PermissionError:
            print(f"Permission denied: {now_path}")
            # 如果遇到权限错误，你可以在这里执行相应的处理操作，或者直接跳过
            pass
     


tf_search_folder = Path("/home/lab4dv/data")
tf_save_folder = Path("/media/lab4dv/新加卷/data_wait_to_be_labeled")
post = Path("sda/duck")
transfer_all_bags_tf_folder(tf_search_folder,tf_save_folder,post)
```
