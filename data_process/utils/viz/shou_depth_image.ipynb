```python
import matplotlib.pyplot as plt
import cv2

depth_image_path = "/media/tony/新加卷/test_data/test/test_1/cam3/depth_to_rgb/image_raw/8.png"
rgb_image_path = "/media/tony/新加卷/test_data/test/test_1/cam3/rgb/image_raw/8.png"

depth_image = cv2.imread(depth_image_path, cv2.IMREAD_UNCHANGED)
rgb_image = cv2.imread(rgb_image_path, cv2.IMREAD_UNCHANGED)
rgb_image = cv2.cvtColor(rgb_image, cv2.COLOR_BGR2RGB)
depth_colormap = cv2.applyColorMap(cv2.convertScaleAbs(
    depth_image, alpha=0.03), cv2.COLORMAP_JET)
depth_colormap[depth_image == 0] = [0, 0, 0]
plt.figure(figsize=(10, 5))
plt.subplot(2, 2, 1)
plt.imshow(depth_image, cmap='gray')
plt.title('Original Depth Image')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(depth_colormap)
plt.title('Colored Depth Image')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(rgb_image)
plt.title('rgb Image')
plt.axis('off')

# Save the resulting imge
# output_image_path = '/mnt/data/colored_depth_image.png'
# cv2.imwrite(output_image_path, depth_colormap)

plt.show()

# Return the path to the saved image
# output_image_path
```
